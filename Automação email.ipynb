```python
# Exportar relatório
import pandas as pd
tabela = pd.read_excel('Vendas_ Dez.xlsx')
display(tabela)
faturamento = tabela['Valor Final'].sum()
Quantidade = tabela['Quantidade'].sum()
print(Quantidade)
print(faturamento)
```

```python
 # enviar um relatório para a diretoria 

import pyperclip as pyp
py.hotkey('ctrl','t')
pyp.copy('https://mail.google.com/mail/u/0/#inbox')
py.hotkey('ctrl','v')
py.press('enter')
time.sleep(3)
py.click(x=112, y=214)
time.sleep(2)
py.write('tidouglas.pinheiro@gmail.com')
py.press('enter')
py.press('tab')
time.sleep(1)
pyp.copy('Relatório de vendas')
py.hotkey('ctrl','v')
py.press('tab')
time.sleep(2)
texto = f"""Prezados, bom dia.
segue o relatório de vendas do último dia;
Quantidade de vendas: {Quantidade}
Faturamento: {faturamento}
Abs.
Douglas Pinheiro
"""
pyp.copy(texto)
py.hotkey('ctrl','v')
py.click(x=964, y=828)
time.sleep(2)
py.write('Vendas_ Dez.xlsx')
py.press('enter')
py.hotkey('ctrl','enter')

    
```
